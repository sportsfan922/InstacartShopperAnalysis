## Create Max Order Column:1st Step of How Many Customers are Loyal, Regular, or New

```python
df_ords_prods_merge['max_order']=df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

## Create Column Based on Conditions: 2nd Step of Sorting Customers into Loyal, Regular, or New

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] >= 10) & (df_ords_prods_merge['max_order'] < 40), 'loyalty_flag'] = 'Regular Customer'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] < 10, 'loyalty_flag'] = 'New Customer'                                              
```

## Getting Basic Stats for Customers with Different Levels of Loyalty

```python
df_ords_prods_merge.groupby('loyalty_flag')['prices'].describe()
```

New Customers, on average, pay the highest price for a product, followed by regular and finally, loyal customers.

```python
df_ords_prods_merge.loc[df_ords_prods_merge['prices']<10, 'spending_flag']='Low Spender'
df_ords_prods_merge.loc[df_ords_prods_merge['prices']>=10, 'spending_flag']='High Spender'
```

```python
## Check Spending Flag Counts
df_ords_prods_merge['spending_flag'].value_counts()
```

## Order Frequency (Places Customers in Groups Based on How Long it Has Been Since They Placed an Order)

```python
df_median=df_ords_prods_merge['days_since_prior_order'].median()
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['days_since_prior_order']>20, 'Order_Flag']='Non Frequent Customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['days_since_prior_order'] > 10) & (df_ords_prods_merge['days_since_prior_order'] <= 20), 'Order_Flag'] = 'Regular Customer'
df_ords_prods_merge.loc[df_ords_prods_merge['days_since_prior_order']<10, 'Order_Flag']= 'Frequent Customer'
```

```python
## Value Counts of Order Frequency Flags
df_ords_prods_merge['Order_Flag'].value_counts()
```

## What is the Median Number of Days That a Customer Waits Between Orders For Frequent Customers (Less than 10 Days), Non Frequent Customers (More than 20 Days), and Regular Customers (Between 11 and 20 Days)

```python
## Regularity of User Ordering Behavior According to Median Method
df_ords_prods_merge.groupby('Order_Flag').agg({'days_since_prior_order': ['median']})
```
